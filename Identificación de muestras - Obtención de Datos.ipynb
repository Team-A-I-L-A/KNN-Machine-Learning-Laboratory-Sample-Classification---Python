```python
mainpath = "drive/My Drive/DATASET/MUESTRAS_SANGRE"

promediosR = np.zeros((86))
promediosG = np.zeros((86))
promediosB = np.zeros((86))
promediosH = np.zeros((86))
promediosS = np.zeros((86))
promediosV = np.zeros((86))
promediosR2 = np.zeros((86))
promediosG2 = np.zeros((86))
promediosB2 = np.zeros((86))
promediosH2 = np.zeros((86))
promediosS2 = np.zeros((86))
promediosV2 = np.zeros((86))
proporcionesR = np.zeros((86))
proporcionesG = np.zeros((86))
proporcionesB = np.zeros((86))
proporcionesH = np.zeros((86))
proporcionesS = np.zeros((86))
proporcionesV = np.zeros((86))
oranges = np.zeros((86))

for i in range(1,87): 
  rutaArchivo = str(i)+".jpg"
  rutaTotal = os.path.join(mainpath,rutaArchivo)

  #OBTENER CAPAS RGB Y HSV DE LA IMAGEN TOTAL
  imagenRGB = cv2.imread(rutaTotal)[...,::-1]
  imgR = imagenRGB[:,:,0].copy()
  imgG = imagenRGB[:,:,1].copy()
  imgB = imagenRGB[:,:,2].copy()

  imagenHSV = cv2.cvtColor(imagenRGB, cv2.COLOR_RGB2HSV)
  imgH = imagenHSV[:,:,0].copy()
  imgS = imagenHSV[:,:,1].copy()
  imgV = imagenHSV[:,:,2].copy()

  #OBTENER CAPAS RGB Y HSV DE LA MITAD IZQUIERDA
  altura, ancho = imgR.shape
  
  imgR2 = imgR[:,0:int(ancho/2)].copy()
  imgG2 = imgG[:,0:int(ancho/2)].copy()
  imgB2 = imgB[:,0:int(ancho/2)].copy()

  imgH2 = imgH[:,0:int(ancho/2)].copy()
  imgS2 = imgS[:,0:int(ancho/2)].copy()
  imgV2 = imgV[:,0:int(ancho/2)].copy()

  #HALLAR PROMEDIOS DE CADA CAPA DE IMAGEN COMPLETA RGB Y HSV
  promedioR = imgR.mean()
  promedioG = imgG.mean()
  promedioB = imgB.mean()
  promedioH = imgH.mean()
  promedioS = imgS.mean()
  promedioV = imgV.mean()

  promediosR[i-1] = promedioR
  promediosG[i-1] = promedioG
  promediosB[i-1] = promedioB
  promediosH[i-1] = promedioH
  promediosS[i-1] = promedioS
  promediosV[i-1] = promedioV

  #HALLAR PROMEDIOS DE CADA CAPA DE MITAD IZQUIERDA RGB Y HSV
  promedioR2 = imgR2.mean()
  promedioG2 = imgG2.mean()
  promedioB2 = imgB2.mean()
  promedioH2 = imgH2.mean()
  promedioS2 = imgS2.mean()
  promedioV2 = imgV2.mean()

  promediosR2[i-1] = promedioR2
  promediosG2[i-1] = promedioG2
  promediosB2[i-1] = promedioB2
  promediosH2[i-1] = promedioH2
  promediosS2[i-1] = promedioS2
  promediosV2[i-1] = promedioV2

  #HALLAR PROMEDIOS DE CADA CAPA DE MITAD IZQUIERDA RGB Y HSV
  proporcionR = promedioR/promedioR2
  proporcionG = promedioG/promedioG2
  proporcionB = promedioB/promedioB2
  proporcionH = promedioH/promedioH2
  proporcionS = promedioS/promedioS2
  proporcionV = promedioV/promedioV2

  proporcionesR[i-1] = proporcionR
  proporcionesG[i-1] = proporcionG
  proporcionesB[i-1] = proporcionB
  proporcionesH[i-1] = proporcionH
  proporcionesS[i-1] = proporcionS
  proporcionesV[i-1] = proporcionV

  if (i<=43):
    orangeIzq = 1
  else:
    orangeIzq = 0
  
  oranges[i-1] = orangeIzq
```

```python
data = pd.DataFrame({'Promedio R Total': promediosR.tolist(),
                     'Promedio G Total': promediosG.tolist(),
                     'Promedio B Total': promediosB.tolist(),
                     'Promedio H Total': promediosH.tolist(),
                     'Promedio S Total': promediosS.tolist(),
                     'Promedio V Total': promediosV.tolist(),
                     'Promedio R Mitad': promediosR2.tolist(),
                     'Promedio G Mitad': promediosG2.tolist(),
                     'Promedio B Mitad': promediosB2.tolist(),
                     'Promedio H Mitad': promediosH2.tolist(),
                     'Promedio S Mitad': promediosS2.tolist(),
                     'Promedio V Mitad': promediosV2.tolist(),
                     'Proporción Promedio R': proporcionesR.tolist(),
                     'Proporción Promedio G': proporcionesG.tolist(),
                     'Proporción Promedio B': proporcionesB.tolist(),
                     'Proporción Promedio H': proporcionesH.tolist(),
                     'Proporción Promedio S': proporcionesS.tolist(),
                     'Proporción Promedio V': proporcionesV.tolist(),
                     'Naranja Izquierda': oranges.tolist()})

data
```
